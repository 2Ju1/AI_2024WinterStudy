```python
#최대 찾아주는 함수 작성
class MyClass:
  def setdata(self, a):
    self.x=a

  def my_max(self):
    temp=self.x
    n=len(self.x)
    current_max=temp[0]
    for i in range(0,n):
      if(current_max<temp[i]):
        current_max=temp[i]
    return current_max
```

```python
p1=MyClass()
p1.setdata([2,4,1,3,7,5])
print(p1.x)
sorted=p1.my_max()
print(sorted)
```

### 1. `init()` Function  
<생성자 생성 방법>
 1. def \_\_init\_\_(self, 매개변수1, ..):
 2. 나머지는 다른 함수와 동일하다(special method라 하는 이유)
 <생성자 사용 방법>
 객체 생성하면서 매개변수 전달
 내부적으로 init() 함수 호출

```python
class Person:
  def __init__(self,x1,x2):
    self.x1=x1
    self.x2=x2

p1=Person("John",[2,5,1,3,4])
print(p1.x1)
print(p1.x2)
```

### 2. `__call__` method
  <사용목적>  
 클래스 객체 만들고 나서 이거를 함수처럼 매개변수 대입하고 쓰고 싶을 때  
 참고) `__init__`는 초기화를 위한 것이어서 처음에만 쓰고 끝이지만 `__call__`은 계속 함수처럼 사용가능  
  <생성 방법>  
  `def __call__(self, 매개변수):`  
  <사용 방법>  
  객체를 함수로 사용하려 시도할 때 내부적으로 실행됨

```python
class Student(Person): #부모 클래스 상속방법
  def __init__(self, x1, x2):
    super().__init__(x1,x2) #부모 클래스 메소드 호출할 땐 super().
  def run(self):
    print(self.x1,self.x2," can run.")
  def __call__self(self):
    return "this is automatically called", self.x1
  def __len__(self):
    print("len 함수")
    return len(self.x1)
```

```python
Me=Student("Jaeyoun","Ahn")
```

```python
Me.run() #그냥 함수 호출
```

```python
Me() #객체를 함수처럼 호출하게 되면 __call__함수 내부 호출
```

```python
Me.__call__() #명시적으로 호출한 결과와 같음
```

```python
len(Me)
```

#### 오버라이딩하기

```python
class Student(Person):
  def __init__(self, fname, lname, year):
    super().__init__(fname, lname)#객체에서 함수 호출하는 느낌이므로 self 안 쓰기
    self.graduationyear=year
  def printname(self):
    super().printname() #자바랑 다른 것 _super뒤에 괄호 붙여야
    print("graduated in",self.graduationyear)
  def play(self):
    print(self.firstname,"play basketball")
```

```python
s=Student("Jaeyoun","Ahn",2012)
s.printname()
s.play()
```

### Example:

Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.
    
Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of ​​the rectangle.
    
Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.
    
    
Create a Parallelepipede child class inheriting from the Rectangle class and with a height attribute and another Volume() method to calculate the volume of the Parallelepiped.


```python
class Rectangle:
  def __init__(self,length, width):
    self.length=length
    self.width=width
  def Perimeter(self):
    self.perimeter=2*(self.length+self.width)
  def Area(self):
    self.area=self.length*self.width

  def display(self):
    print("length:",self.length,"width:",self.width, "Perimeter:",self.perimeter,"Area:",self.area)
```

```python
R1=Rectangle(2,3)
R1.Perimeter()
R1.Area()
R1.display()
```

```python
class Parallelepipede(Rectangle):
  def __init__(self, length, width, height):
    super().__init__(length, width)
    self.height=height
  def Volume(self):
    super().Area()
    self.volume=self.area*self.height
  def display(self):
    self.Volume()
    print("length:",self.length,"width:",self.width,"height:",self.height,"volume:",self.volume)
```

```python
p1=Parallelepipede(2,3,4)
p1.display()
```
