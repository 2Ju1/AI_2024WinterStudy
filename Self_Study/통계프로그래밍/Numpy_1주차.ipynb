## 1. Basic Operations

````python
import numpy as np

storages = [24, 3, 4, 23, 10, 12] // 파일형식 list
print(storages)

np_storages = np.array(storages)

print(type(np_storages))
````

````python
# [1, 2, 3, 4] + [1, 2, 3, 4] without using Numpy
storages = [1, 2, 3, 4]
new_storages = []
for n in storages:
    n += n
    new_storages.append(n)
print(new_storages)
````

````python
# [1, 2, 3, 4] + [1, 2, 3, 4] using Numpyimport numpy as np
storages = np.array([1, 2, 3, 4])
print(id(storages))
storages += storages
print(storages)
print(id(storages))

storages = storages+storages
print(id(storages)) # id has changed
````

Meaning that you do not have to specify an exact number for one of the dimensions in the reshape method.

Pass -1 as the value, and NumPy will calculate this number for you.

## View VS Copy

The main difference between a copy and a view of an array is that the copy is a new array, and the view is just a view of the original array.

````python
# Alias (View)

arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)


arr2 = arr1
arr2[0] = 100

print(arr1)

````

````python
# Copy: copy the materials only.
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

arr2 = arr1.copy()
arr2[0] = 100

print(arr2)
````

## Comparison with List

List: slicing will do copy

Numpy: slicing will do view

````python
import numpy as np


arr_List = [x for x in range(10)]
print(arr_List)

arr_List_copy = arr_List[:-4]
arr_List_copy[0] = 100
print(arr_List_copy)
print(arr_List)
````

````python
arr_NumPy = np.arange(10)
print(arr_NumPy)


arr_NumPy_view = arr_NumPy[:-4]
arr_NumPy_view[0] = 100

print(arr_List_copy)
print(arr_NumPy)

````

````python
# Recommendation: use .copy() in Numpy.
arr_NumPy = np.arange(10)
arr_NumPy_copy = arr_NumPy[:-4].copy()
arr_NumPy_copy[0] = 100
print(arr_NumPy_copy)
print(arr_NumPy)
````

### Universal Functions

````python
import numpy as np

arr = np.array([4, -9, 16, -4, 20])
print(arr)

arr_abs = np.abs(arr)
print(arr_abs)


print(np.exp(arr_abs))
print(np.sqrt(arr_abs))
````

````python
import numpy as np

arr1 = [2, 5, 7, 9, 5, 2]
arr2 = [2, 5, 8, 3, 1]

new_arr1 = np.unique(arr1)
print(new_arr1)

print(np.union1d(new_arr1, arr2))

print(np.intersect1d(new_arr1, arr2))

print(np.setdiff1d(new_arr1, arr2))
````

### axis

Function will be applied along with axis = 0. As a result, axis = 0 will be summarized.

````python
arr = np.array([[1, 2 ,3], [4, 5, 6]])
print(arr.shape) #axis 0 means the first element "2" out of (2,3).
print(arr.sum())
print(arr.sum(axis=0))    # array 형태가 아니라
#print(arr.sum(axis=1))
````

###  Rule of Broadcasting
Rule 1: If number of dim does not match, the shape of one is padded on the left side.

Rule 2: Assume that number of dim match. If shape does not match, the array with shape 1 is stretcted.

Rule 3: If in any dimension the sizes disagree and neither is equal to 1, an error is raised.
