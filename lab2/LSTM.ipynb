## input의 차원이 1인 경우

```python
class LSTMCell_dim1(nn.Module) :
    def __init__(self) :
        super(LSTMCell_dim1, self).__init__()
        self.ii=nn.Linear(1,1)
        self.hi=nn.Linear(1,1)

        self.If=nn.Linear(1,1)
        self.hf=nn.Linear(1,1)

        self.ig=nn.Linear(1,1)
        self.hg=nn.Linear(1,1)

        self.io=nn.Linear(1,1)
        self.ho=nn.Linear(1,1)

        self.sigmoid=nn.Sigmoid()
        self.tanh=nn.Tanh()

    def forward(self, xt, h_t_1,c_t_1) :
        # xt, h_t_1, c_t_1은 1차원 scalar
        #calculate ht, ct
        xt=torch.flatten(xt)
        ht_1=torch.flatten(ht_1)
        it=self.sigmoid(self.ii(xt)+self.hi(h_t_1))
        ft=self.sigmoid(self.If(xt)+self.hf(h_t_1))
        gt=self.tanh(self.ig(xt)+self.hg(h_t_1))
        ot=self.sigmoid(self.io(xt)+self.ho(h_t_1))

        ct=ft*c_t_1+it*gt
        ht=ot*self.tanh(ct)
        return ht, ct
```

```python
class LSTMModel_dim1(nn.Module):
    def __init__(self) :
        super(LSTMModel_dim1, self).__init__()
        self.lstm = LSTMCell_dim1()
        self.fc=nn.Linear(1,1)#fully connected layer(마지막 MLP)

        self.optimizer = optim.SGD(self.parameters(), lr=0.01)

    def forward(self, x) :
        # 이때 X는 1차원 vector
        cn=torch.ones(1,1,dtype=torch.float32)
        hn=torch.ones(1,1,dtype=torch.float32)
        outs_list=[]
        for seq in range(len(x)) :
            hn, cn = self.lstm(x[seq], hn, cn) # x[seq]를 1차원 텐서로 만듦
            outs.append(self.fc(hn))
        outs=torch.cat(outs_list, dim=0)
        return outs
        #backward할 때 outs의 각 요소별 loss를 알아야 함/



    def predict(self,x,num):#실제 값을 넣어 예측하는 상황
        cn=torch.ones(1,1,dtype=torch.float32)
        hn=torch.ones(1,1,dtype=torch.float32)
        prediction=[]

        for seq in range(len(x)) :
            hn, cn = self.lstm(x[seq], hn, cn)

        x=self.fc(hn)#마지막 상태에서의 예측값
        prediction.append(x)


      #내가 아는 길이까지만 내용을 넣어주기

        #기존의 결과들이랑 마지막의 결과까지 합쳐주기
        for i in range(num): #내가 알고 싶은 개수 만큼
            hn,cn = self.lstm(x,hn,cn)
            x=self.fc(hn)
            prediction.append(x)
        #prediction=torch.cat(prediction, dim=0)
        return prediction
        #이 prediction 배열 안에는 내가 예측시작한 첫번째 단어부터 리스트의 형태로 저장
```
