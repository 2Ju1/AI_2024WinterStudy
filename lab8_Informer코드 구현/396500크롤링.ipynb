```python
# yfinance 라이브러리 설치
!pip install yfinance

# 라이브러리 임포트
import yfinance as yf

# 주식 데이터 가져오기
tickerSymbol = '091230.KS' # 예시로 Apple Inc.의 주식 심볼 사용
tickerData = yf.Ticker(tickerSymbol)

# 지정된 기간 동안의 주식 데이터 가져오기
tickerDf = tickerData.history(period='1d', start='2018-01-01', end='2022-12-31')

# 데이터 출력
print(tickerDf)
```

```python
tickerDf.to_csv('/content/drive/MyDrive/study_lab/lab8_Informer코드 구현/091230.csv', index=True)
```
